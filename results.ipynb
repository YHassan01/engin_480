```python
# Eystrasalt Offshore Wind Farm â€“ Boundary Visualizer

#This notebook demonstrates how to extract and visualize the boundary of the Eystrasalt offshore wind farm using Python. The code uses a reusable class to convert boundary coordinates into a NumPy array and plots the polygon using Matplotlib.

    
```

```python
import numpy as np
import matplotlib.pyplot as plt

class WindFarmBoundary:
    def __init__(self, name, coordinates):
        self.name = name
        self.coordinates = coordinates
        self.array = np.array(coordinates)

    def get_array(self):
        return self.array

    def plot(self):
        x, y = self.array[:, 0], self.array[:, 1]
        plt.figure(figsize=(6, 6))
        plt.plot(x, y, marker='o', linestyle='-', color='blue')
        plt.fill(x, y, color='skyblue', alpha=0.3)
        plt.title(f"Boundary of {self.name}")
        plt.xlabel("Longitude")
        plt.ylabel("Latitude")
        plt.grid(True)
        plt.axis("equal")
        plt.show()
```

```python
eystrasalt_coords = [
    (17.004, 57.008),
    (17.005, 57.005),
    (17.005, 57.001),
    (17.005, 56.998),
    (17.006, 56.994),
    (17.004, 57.008)  # close the polygon
]

eystrasalt = WindFarmBoundary("Eystrasalt", eystrasalt_coords)
```

```python
eystrasalt.plot()
```
